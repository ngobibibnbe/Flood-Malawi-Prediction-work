# parmeters with Normalization

```python
def findtheta(X,Y):
    inv=np.dot(X.T,X)
    inv=np.linalg.inv(inv)
    theta=(np.dot(inv,X.T)).dot(Y)
    theta=theta.reshape((X.shape[1],1))
    return theta
```

# loss

```python
def losss(theta,X,Y):
    J=0
    for i in range(X.shape[0]):
        #print(X[i,:])
        J+=(np.dot(theta.T,X[i,:].T)-Y[i])*(np.dot(theta.T,X[i,:].T)-Y[i])
        J=J
    return np.squeeze(J)
```
