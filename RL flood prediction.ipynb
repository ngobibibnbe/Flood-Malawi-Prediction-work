# parmeters with Normalization

```python
def findtheta(X,Y):
    inv=np.dot(X.T,X)
    inv=np.linalg.inv(inv)
    theta=(np.dot(inv,X.T)).dot(Y)
    return theta
```

# loss

```python
(n,m)=X.shape
def losss(theta,X,Y):
    J=0
    for i in range(n):
        J+=(np.dot(theta.T,X[i,:])-Y[i])*(np.dot(theta.T,X[i,:])-Y[i])
    J=J/(2*n)
    return J
n
```

```python
theta=findtheta(X,Y)
losss(theta)
```
